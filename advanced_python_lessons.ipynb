### Combining Series & DataFrames
- 'concat()' -Allows us to add together DF of similar shape
- 'append()'
- 'merge()' - similar to JOIN in SQL
- .join()

```python
import pandas as pd
import numpy as np
```

```python
# .concat()
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },

)


df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },

)


df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },

)

frames = [df1, df2, df3]

# ignore_index to create a new index starting from 0 - n
result = pd.concat(frames, ignore_index=True)

result
```

```python
# use keys to delineate the diff concatenated DF
# allows tracking of original DF
result = pd.concat(frames, keys=['x', 'y', 'z'])

result
```

```python
# Allows extraction of the original DF objects

result_y = result.loc[('y')]
result_y
```

```python
df4 = pd.DataFrame(
    {
        "B": ["B2", "B3", "B6", "B7"],
        "D": ["D2", "D3", "D6", "D7"],
        "F": ["F2", "F3", "F6", "F7"],
    },
    index=[12, 13, 14, 15],
)

df4
```

```python
# if our df have mismatched columns, NAn is filled in where they donot overlap
result = pd.concat([df1, df4], ignore_index=True, sort=False)
result
```

```python
result = pd.concat([df1, df4], ignore_index=True, axis=1, sort=False)
result
```

### pd.merge() -similar to Join SQL. Combining on index or column

```python
left_df = pd.DataFrame({'Courses': ["Spark","PySpark","Python","pandas","Java"],
                    'Fee' : [20000,25000,30000,24000,40000],
                    'Duration':['30day','40days','60days','55days','50days']})

right_df = pd.DataFrame({'Courses': ["Java","PySpark","Python","pandas","Hyperion","html"],
                    'Fee': [20000,25000,30000,24000,40000,4000],
                    'Percentage':['10%','20%','25%','20%','10%','50%']})
```

```python
left_df
```

```python
right_df
```

```python
merged_df = pd.merge(left_df, right_df)
merged_df
```

```python
merged_df = pd.merge(left_df, right_df, on='Courses')
merged_df
```

```python
merged_df = pd.merge(left_df, right_df, on=['Courses', 'Fee'])
merged_df
```

```python
merged_df = pd.merge(left_df, right_df, how='left', left_on=['Courses', 'Fee'], right_on=['Courses','Fee'])
merged_df
```

```python
left = pd.DataFrame({"A": [1, 2], "B": [1, 2]})

right = pd.DataFrame({"A": [4, 5, 6], "B": [2, 2, 2]})
```

```python
result = pd.merge(left, right, on='B', how='inner')
result
```

```python
result = pd.merge(left, right, on='B', how='inner', validate='one_to_many')
result
```

### pd.join()

```python
technologies = {
    'Courses':["Spark","PySpark","Python","pandas"],
    'Fee' :[20000,25000,22000,30000],
    'Duration':['30days','40days','35days','50days'],
              }
index_labels=['r1','r2','r3','r4']

technologies2 = {
    'Courses':["Spark","Java","Python","Go"],
    'Discount':[2000,2300,1200,2000]
              }
index_labels2=['r1','r6','r3','r5']
```

```python
df1 = pd.DataFrame(technologies, index=index_labels)
df2 = pd.DataFrame(technologies2, index=index_labels2)
```

```python
# left join
df3 = df1.join(df2, lsuffix='_left', rsuffix='_right')
df3
```

```python
# inner join
df3 = df1.join(df2, lsuffix='_left', rsuffix='_right', how='inner')
df3
```

```python
# right joins
df3 = df1.join(df2, lsuffix='_left', rsuffix='_right', how='right')
df3
```

```python
# joining df setting indexes
df3 = df1.set_index('Courses'.join(df2.set_index('Courses'), how='inner'))
df3
```

## Group by with student scores

```python
df = pd.read_csv('data/student_scores.csv', parse_dates=['birth'])

rows, cols =df.shape
print(f'Rows: {rows}')
print(f'Cols: {cols}')

df.describe()
```

```python
df.info()
```

```python
df.head()
```

```python
item_group = df.groupby('first_name')
item_group.groups
```

```python
multi_group = df.groupby(['first_name', 'last_name'])
multi_group.groups
```

```python
# looping/iterating through groups
for name, group in item_group:
    print('{}' .format(name))
    print(group)
```

```python
subject_group = df.groupby('Subject')
subject_group.groups

# .get_group() to then grab a specific group
subject_group.get_group('Calculus')
```

```python
agg_group_mean = df.groupby('Subject')['score'].mean()
agg_group_mean
```

```python
student_means = df.groupby(['first_name', 'last_name'])['score'].mean()
print(student_means)
```

```python
agg_group = df.groupby(['first_name', 'last_name'])['score'].agg([np.mean, np.sum, np.median])
print(agg_group)
```

```python
student_count = df.groupby(['first_name', 'last_name'])['id'].count()
student_count
```

### .pivot() - reshaping data based on columns

```python
data = {
    'Date': ['2022-01-01', '2022-01-01', '2022-01-02', '2022-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Value': [10, 20, 15, 25]
}

df = pd.DataFrame(data)
df
```

```python
pivot_df = df.pivot(index='Date', columns='Category', values='Value')
pivot_df
```

```python
data = {'Name': ['Alice', 'Bob', 'Alice', 'Charlie'],
        'Fruit': ['Apple', 'Banana', 'Orange', 'Apple'],
        'Quantity': [3, 2, 1, 4]}

df = pd.DataFrame(data)
df
```

```python
pivot = df.pivot(index='Name', columns='Fruit', values='Quantity')
pivot
```

```python
import datetime
```

```python
# Create a DataFrame with daily returns for different stocks
np.random.seed(42)
start_date = datetime.date(2022, 1, 1)
end_date = datetime.date(2022, 1, 10)
date_range = pd.date_range(start_date, end_date, freq='D')
# Create a DataFrame with daily returns for different stocks
data = {
    'Date': np.repeat(date_range, 3),
    'Stock': ['AAPL', 'GOOGL', 'MSFT'] * len(date_range),
    'Return': np.random.normal(loc=0.001, scale=0.01, size=len(date_range) * 3)
}

returns_df = pd.DataFrame(data)
```

```python
returns_df
```

```python
pivot = returns_df.pivot(index='Date', columns='Stock', values='Return')
pivot
```
