### 386.4.4 R-GLAB Aggregate Functions

```python
import pandas as pd
```

```python
data = {
    'Category': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Value': [10, 15, 20, 25, 30, 35],
    'Fee' :[20000,25000,26000,22000,24000,35000],
    'Duration':['30day','40days','35days','40days','60days','60days'],
    'Discount':[1000,2300,1200,2500,2000,2000]
}

df = pd.DataFrame(data)
```

```python
df.head(6)
```

```python
total_sum =df['Value'].aggregate('sum')
print('Total Sum:', total_sum)

total_sum2 = df['Value'].sum()
print(f'Total Sum: {total_sum2}')
```

```python
# Cumulative Sum (Running Total)
df[['Value']].cumsum()
```

```python
total_mean = df['Value'].aggregate('mean')
print(f'Total mean: {total_mean}')

mean = df['Value'].mean()
print(f'Mean: {mean}')
```

```python
max_val = df['Value'].agg('max')
print(f'Max Value: {max_val}')

max = df['Value'].max()
print(f'Max Value: {max}')
```

```python
count = df['Value'].agg('count')
print(f'Count: {count}')

count1 = df['Value'].count()
print(f'Count: {count1}')
```

```python
std = df['Value'].agg('std')
print(f'Standard Dev: {std}')

std1 = std = df['Value'].std()
print(f'Standard Dev: {std1}')
```

## Apply Multiple Aggregation Functions

* 'df[] - grabbing a series from a df - single column
* 'df[[]] - grabbing a dataframe from a df - multicolumn

```python
result = df[['Fee', 'Discount']].aggregate('sum')

print(result)

result2 = df[['Fee', 'Discount']].sum()
result2
```

```python
# Determine the total of eachg group with and agg function
result_group = df.groupby('Category')[['Fee', 'Discount']].agg('sum')

print(result_group)
```
