# Concatenating from separate Files

```python
import pandas as pd
```

```python
re_df1 = pd.read_csv('data/RealEstate1.csv')
re_df2 = pd.read_csv('data/RealEstate2.csv')
re_df3 = pd.read_csv('data/RealEstate3.csv')
```

```python
realEstate = pd.concat([re_df1, re_df2, re_df3], axis=0, ignore_index=False)
realEstate
```
